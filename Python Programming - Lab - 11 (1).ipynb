### 01) WAP to demonstrate the use of Pie chart.

```python
x=[10,20,30,40]
y=['food','travel','accomodation','shopping']
plt.pie(x,labels=y,colors=['r','g','b','y'],explode=[0,0,0,1],shadow=True,radius=1.2,autopct="%0.1f%%",labeldistance=1.2)
plt.title('hghg')
plt.show()
```

### 02) WAP to to Plot List random of X, Y Coordinates in Matplotlib.

### 03) WAP to demonstrate the use of Bar chart.

```python
x=['python','c','c++','java']
y=[85,70,60,30]
plt.barh(x,y.color=['r','g','b','y'],edgecolor='b',alpha=0.5)
plt.title="popularity of language"
plt.xlabel("lag")
plt.ylabel("popularity")
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x=['python','c','c++','java']
y=[85,70,60,30]
plt.bar(x,y,color=['r','g','b','y'],edgecolor='b',alpha=0.5)
plt.title="popularity of language"
plt.xlabel("lang")
plt.ylabel("popularity")
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
addlabels(x,y)        
plt.show()
```
